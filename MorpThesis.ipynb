```python
# to drop the members who attempt gaming the system, I crate a deep copy of exams_df to exams_df_nongts.

exams_df_nongts = exams_df.copy(deep=True)

# create a new column for labeling and overlapping time

exams_df_nongts['Label'] = ""
exams_df_nongts['Overlap (secs.)'] = ""


# then we write the conditions for labeling the data who "Gaming the System"

for i in range(1,len(exams_df_nongts)):
    if exams_df_nongts['Üye'][i-1] == exams_df_nongts['Üye'][i]:
        if exams_df_nongts['Sınav'][i-1] == exams_df_nongts['Sınav'][i]:
            if exams_df_nongts['Konu'][i-1] == exams_df_nongts['Konu'][i]:
                if (exams_df_nongts['Çıkış Zamanı'][i-1] > exams_df_nongts['Giriş Zamanı'][i]) and (exams_df_nongts['Puan'][i-1] < exams_df_nongts['Puan'][i]) and (exams_df_nongts['GirişLog'][i-1] == exams_df_nongts['GirişLog'][i]):
                    exams_df_nongts['Label'][i] = "Gaming the System"
                    exams_df_nongts['Overlap (secs.)'][i] = (exams_df_nongts['Giriş Zamanı'][i] - exams_df_nongts['Çıkış Zamanı'][i-1]).total_seconds() * -1  # this is the overlapping time in seconds 
```

```python
#exam data without gaming the system incidents

exams_df_nongts=exams_df_nongts.loc[exams_df_nongts['Label'] != 'Gaming the System']

exams_df_nongts.sort_values(by=['Konu','Üye'])

exams_df_nongts
```

```python
#get Puan and Süre values and assign to x

x=df.iloc[:,[3,7]].values
```

```python
#finding optimal number of clusters using the elbow method  
import matplotlib.pyplot as mtp   

from sklearn.cluster import KMeans
wcss_list=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
mtp.plot(range(1,11),wcss_list)
mtp.title('The Elbow Method')
mtp.xlabel('Number of clusters')
mtp.ylabel('WCSS')
mtp.show()    

```

```python
#training the K-means model on a dataset  

kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42)
y_predict=kmeans.fit_predict(x)
```

```python
#visulaizing the clusters

mtp.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
mtp.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
mtp.scatter(x[y_predict== 2, 0], x[y_predict == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
mtp.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster  
#mtp.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') #for fifth cluster
#mtp.scatter(x[y_predict == 5, 0], x[y_predict == 5, 1], s = 100, c = 'purple', label = 'Cluster 6') #for sixth cluster    
mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')   
mtp.title('Clusters of students')  
mtp.xlabel('Puan')  
mtp.ylabel('Süre')  
mtp.legend()  
mtp.show()  
```
